```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\panav\Desktop\ML\Datasets\penguins_lter.csv')
data
```

```python
data['studyName'] = pd.factorize(data['studyName'])[0]
data
```

```python
data['Species'] = pd.factorize(data['Species'])[0]
data['Region'] = pd.factorize(data['Region'])[0]
data['Island'] = pd.factorize(data['Island'])[0]
data['Stage'] = pd.factorize(data['Stage'])[0]
data['Clutch Completion'] = pd.factorize(data['Clutch Completion'])[0]
data['Date Egg'] = pd.factorize(data['Date Egg'])[0]
data['Sex'] = pd.factorize(data['Sex'])[0]
data
```

```python
data['Culmen Length (mm)'] = data['Culmen Length (mm)'].fillna(data['Culmen Length (mm)'].mean())
data
```

```python
data['Culmen Depth (mm)'] = data['Culmen Depth (mm)'].fillna(data['Culmen Depth (mm)'].mean())
data
```

```python
data['Flipper Length (mm)'] = data['Flipper Length (mm)'].fillna(data['Flipper Length (mm)'].mean())
data['Body Mass (g)'] = data['Body Mass (g)'].fillna(data['Body Mass (g)'].mean())
data['Delta 15 N (o/oo)'] = data['Delta 15 N (o/oo)'].fillna(data['Delta 15 N (o/oo)'].mean())
data['Delta 13 C (o/oo)'] = data['Delta 13 C (o/oo)'].fillna(data['Delta 13 C (o/oo)'].mean())
data
```

```python
data = data.drop('Comments', axis = 1)
data = data.drop('Individual ID', axis = 1)
data
```

```python
data = data[[col for col in data.columns if col != 'Species' ] + ['Species']]
data.head()
```

```python
data
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(data.corr(), annot = True)
```

```python
data = data.drop('Region', axis = 1)
data = data.drop('Stage', axis = 1)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:, :13]
y = data['Species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train, y_train)
```

```python
pred = lg.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
acc_lg = accuracy_score(pred, y_test)
acc_lg
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
pred_svm = clf.predict(X_test)
pred_svm
```

```python
acc_svm = accuracy_score(pred_svm, y_test)
acc_svm
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
pred_dtc = dtc.predict(X_test)
pred_dtc
```

```python
acc_dtc = accuracy_score(pred_dtc, y_test)
acc_dtc
```

**Hence, it can be clearly seen that Decision Tree Classifier has performed best for this dataset (100% accuracy). Logistic Regression also gave satisfactory accuracy (94.23%), with SVM giving the least accuracy (74.04%).**
